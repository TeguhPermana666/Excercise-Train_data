```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
iris = load_iris()
list(iris.keys())
```

```python
print(iris.DESCR)
```

```python
X=iris["data"][:,(2,3)] # petal length and width
# y = (iris["target"]==2).astype(np.int) # iris virginicia
y = iris['target']# take all of Y
```

```python
soft_max_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs',C=10,random_state=42)
soft_max_reg.fit(X,y)
# multi class = multinomial akan menarik data ke dalam ovr methode dalam multiclass
# yang semula yakno OVO dalam binary classifier
```

```python
# be know how to match the score di setiap class label
print(soft_max_reg.predict([[5,2]]))
print(soft_max_reg.predict_proba([[5,2]]))
```

jadi dapat di lihat bahwa dari hasil predic probanya di dapat bahwa score pada index ke dua terbesar yang mana dpt di tarik kesimpulan bahwa dia memiliki nilai condongg ke iris virginicia

```python
x0, x1 = np.meshgrid(
    np.linspace(0,8,500).reshape(-1,1),
    np.linspace(0,3.5,200).reshape(-1,1)
)
```

```python
X_new=np.c_[x0.ravel(),x1.ravel()]
y_proba = soft_max_reg.predict_proba(X_new)
y_predict = soft_max_reg.predict(X_new)
```

```python
list(y_proba)
```

```python
list(y_predict)
```

```python
zz1 = y_proba[:,1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)
```
